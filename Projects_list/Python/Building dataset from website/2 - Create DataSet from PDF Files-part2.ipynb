**So, before merging the dataframes, i searched for an option to keep the dataframe name as a column to know what dataframe  belonged to every data. That gave also the chance to split it and create 2 columns for "unit audit" and the "year".** 

**In order to do that, i created a tuple with the dataframes names to let them as keys and get an index. But after applying reset_index, it rested as a columns**

```python
#LIST OF ALL DATAFRAMES CREATED

frames = [fp_2015, fp_2016, fp_2017,
 fp_2018,  fp_2019,  fp_2020, res_FirmPrivs_2016,
 res_FirmPrivs_2017,  res_FirmPrivs_2018,
 res_FirmPrivs_2019, res_FirmPrivs_2020,
 res_uai_2015, res_uai_2016,
res_uai_2017,res_uai_2018,
 res_uai_2019,  res_uai_2020]
```

```python
# TUPLES OF DATAFRAMES TO ASSIGN THEM AS INDEX

keys = ('fp_2015', 'fp_2016', 'fp_2017',
 'fp_2018',  'fp_2019',  'fp_2020', 'res_FirmPrivs_2016',
 'res_FirmPrivs_2017',  'res_FirmPrivs_2018',
 'res_FirmPrivs_2019', 'res_FirmPrivs_2020',
 'res_uai_2015', 'res_uai_2016',
'res_uai_2017','res_uai_2018',
 'res_uai_2019',  'res_uai_2020')
```

```python
# CONCAT THE DATAFRAMES, ADDING reset_index TO CONVERT INDEX INTO COLUMNS

res_audits_2015_2020 = pd.concat(frames, keys = keys).reset_index()
res_audits_2015_2020
```
