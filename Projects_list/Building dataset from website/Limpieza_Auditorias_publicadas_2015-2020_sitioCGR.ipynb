```python
# IMPORT NECESSARY LIBRARIES

import pandas as pd
import numpy as np
from fuzzywuzzy import process, fuzz
```

```python
# OPEN THE FILE AND STORE IN OUR DATAFRAME

audits_15_20 = pd.read_csv('/Users/leona/Documents/Python/Contraloria/resol_publics_audits_15-20.csv', index_col=False)
audits_15_20
```

## As i said, we're going to need to get a list with the values/names more accurate as possible. We have that information in that data set but it's a lot.

**The  first thing to do is get a dataframe to know the match_score for every name. In this case it's going to get a dataframe with 3 column: the original column, the match and the score. The more close to 100 the more similar is. So, with process.extract it is created a tuple with this information, which is going to be converted as DataFrame**

```python
score_sort = [(x,) + i
             for x in audits_15_20['institucion'].values 
             for i in process.extract(x, audits_15_20['institucion'].values , scorer=fuzz.token_set_ratio)]
```

```python
#Create a dataframe from the tuples

similarity_sort = pd.DataFrame(score_sort, columns=['inst_sort','match_sort','score_sort'])
similarity_sort.sort_values(by=['score_sort'])
```

```python
# WE SELECT ONLY THE DATA WHERE SCORE_SORT IS EQUAL TO 100

higher_match = similarity_sort[similarity_sort['score_sort'] == 100 ]
higher_match
```

**Now, i will check for the names in the match_sort columns to know the more accurate names of the "institucion" column**

```python
higher_match_list = higher_match.match_sort.sort_values().unique().tolist()
print(higher_match_list)
```

```python
print("The number of Institutucion now is:",higher_match.match_sort.nunique() )
```

### After running the code above, we can check that there are still some similar names like "asamblea" and "AsambleaNacional". The number of "institucion" reduced to 359 from 381

### The next step is to add a new column with the best matchs for institutions. To do that, the below code is necessary.

```python
# Create a function in order to store the new values

institucion_bien=[]
proporcion = []

def cambios(mal, bien):
    for i in mal:
        x=process.extractOne(i, bien)
        institucion_bien.append(x[0])
        proporcion.append(x[1])
    return institucion_bien, proporcion

mal = audits_15_20['institucion'].values
bien = higher_match_list


corregido, rango = cambios(mal, bien)
```

```python
# ADD NEW COLUMNS WITH NEW VALUES "inst_buenas" and "ratiom", in order to compare how was matched the
# "institucion" column

audits_15_20['inst_buenas'] = pd.Series(corregido)
audits_15_20['ratiom'] = pd.Series(rango)

audits_15_20
```
